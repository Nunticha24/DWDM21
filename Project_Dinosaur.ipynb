###ตรวจสอบว่าข้อมูลซ้ำกันหรือไม่

าค่า min max ของ column NO.

```python
def min_max(i):
    a = min(i['No.'])
    b = max(i['No.'])
    print(f'min of NO. is {a}')
    print(f'max of NO. is {b}')
```

ดูค่า min max ของ covid ทั้ง 4 ตาราง

```python
min_max(data_covid2020)
```

```python
min_max(data_covid2021_1)
```

```python
min_max(data_covid2021_2)
```

```python
min_max(data_covid2022)
```

เห็นได้ว่าลักษณะของข้อมูลที่ดูจากค่าในคอลัมน์ NO. ของตาราง data_covid2020 และ data_covid2021_1 มีค่าซ้ำกัน

###ดูลักษณะข้อมูลว่าซ้ำกันจริงหรือไม่

```python
data_covid2020.iloc[816989:,:].head(10)    #ดูข้อมูล data_covid2020 ตั้งแต่แถวที่ 816989 ถึงแถวสุดท้าย
```

```python
data_covid2021_1.head(10)    #ดูข้อมูล data_covid2021_1
```

จากการดูลักษณะข้อมูล จะเห็นว่าข้อมูลซ้ำกันจริง

จึงทำการลบข้อมูลที่ซ้ำออก โดยเลือกจากตาราง data_covid2020

```python
data_covid2020_drop = data_covid2020.iloc[:816989,:]
data_covid2020_drop.head()
```

```python
min_max(data_covid2020_drop)   #ดู No. น้อยสุดและมากสุด ใน data_covid2020_drop
```

```python
min_max(data_covid2022) #ดู No. น้อยสุดและมากสุด ใน data_covid2022
```

```python
data_covid2020_drop.shape[0] + data_covid2021_1.shape[0] + data_covid2021_2.shape[0] + data_covid2022.shape[0]
```
